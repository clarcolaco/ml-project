```python
import pandas as pd
```

```python
df = pd.read_csv('pizzas.csv')
```

```python
df
```

```python
df.plot(kind="scatter",x="diametro",y="preco");
```

```python
#### regressao linear
from sklearn import linear_model
modelo = linear_model.LinearRegression()

#### dados de treino
x = df[["diametro"]]
y = df[["preco"]]

##### treino
modelo.fit(x,y)
```

```python
#### previsao para 30 de diametro, 130 e 60
######
array_30_dim = modelo.predict([[30]])

#### pizza 130 diametro
array_130_dim = modelo.predict([[130]])

#### pizza 1000 diametro
array_1000_dim = modelo.predict([[1000]])

#### pizza de 60
array_60_dim = modelo.predict([[60]])
```

```python
### previsao
print(f'Para pizza de 40 diametro valor: {round(float(array_30_dim[0][0]))} reais')
print(f'Para pizza de 130 diametro valor: {round(float(array_130_dim[0][0]))} reais')
print(f'Para pizza de 60 diametro valor: {round(float(array_60_dim[0][0]))} reais')
print(f'Para pizza de 1000 diametro valor: {round(float(array_1000_dim[0][0]))} reais')
```
